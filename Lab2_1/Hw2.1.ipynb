9. Замініть відсутні дані в стовпцях середніми значеннями по цим стовпцям (метод `fillna`).

```python
import pandas as pd
import numpy as np

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"

try:
    dfs = pd.read_html(url)

    target_df = None
    for df in dfs:
        if 'Регіон' in df.columns and '1950' in df.columns and '2014' in df.columns:
             if 'Україна' in df['Регіон'].values:
                target_df = df
                break

    if target_df is not None:
        print("Успішно знайдено таблицю 'Коефіцієнт народжуваності'.")

        target_df = target_df.replace(["—", "-"], np.nan)
        columns_to_convert = target_df.columns.drop('Регіон')
        for col in columns_to_convert:
            target_df[col] = pd.to_numeric(target_df[col], errors='coerce')

        if not target_df.empty:
             target_df = target_df.drop(target_df.index[-1])
             print("\nОстанній рядок (дані по Україні) видалено.")

        print("\nПеревірка наявності NaN ДО заповнення:")
        print(target_df.isnull().sum()) 

        print("\nЗаміна відсутніх даних середніми значеннями по стовпцях...")

        numeric_cols = target_df.select_dtypes(include=np.number).columns

        column_means = target_df[numeric_cols].mean()

        target_df.fillna(column_means, inplace=True)

        print("Заповнення NaN середніми значеннями завершено.")
        print("\nПеревірка наявності NaN ПІСЛЯ заповнення:")
        print(target_df.isnull().sum())

        print("\nПерші 5 рядків ПІСЛЯ заповнення NaN:")
        print(target_df.head())

    else:
        print(f"Помилка: Не вдалося знайти таблицю з очікуваними стовпцями на сторінці.")
        print(f"Всього знайдено таблиць: {len(dfs)}")

except Exception as e:
    print(f"Сталася помилка під час читання або обробки сторінки: {e}")
```

10. Отримайте список регіонів, де рівень народжуваності в 2019 році був вище середнього по Україні.

```python
import pandas as pd
import numpy as np

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
YEAR_COLUMN = '2019' 

try:
    dfs = pd.read_html(url)

    target_df = None
    for df in dfs:
        if 'Регіон' in df.columns and '1950' in df.columns and YEAR_COLUMN in df.columns:
             if 'Україна' in df['Регіон'].values:
                target_df = df
                break

    if target_df is not None:
        print("Успішно знайдено таблицю 'Коефіцієнт народжуваності'.")

        target_df = target_df.replace(["—", "-"], np.nan)

        columns_to_convert = target_df.columns.drop('Регіон')
        for col in columns_to_convert:
            if col in target_df.columns:
                 target_df[col] = pd.to_numeric(target_df[col], errors='coerce')

        if not target_df.empty and 'Україна' in target_df['Регіон'].iloc[-1]: 
             target_df = target_df.drop(target_df.index[-1])
             print("\nОстанній рядок (дані по Україні) видалено.")

        numeric_cols = target_df.select_dtypes(include=np.number).columns
        if not target_df.empty:
             column_means = target_df[numeric_cols].mean()
             target_df.fillna(column_means, inplace=True)
             print("Відсутні дані заповнено середніми.")

        if YEAR_COLUMN in target_df.columns:
            average_birth_rate_2019 = target_df[YEAR_COLUMN].mean()
            print(f"\nСередній коефіцієнт народжуваності по регіонах у {YEAR_COLUMN} році: {average_birth_rate_2019:.2f}")

            regions_above_average_df = target_df[target_df[YEAR_COLUMN] > average_birth_rate_2019]

            regions_list = regions_above_average_df['Регіон'].tolist()

            if regions_list:
                print(f"\nРегіони, де народжуваність у {YEAR_COLUMN} році була вище середньої:")
                for region in regions_list:
                    print(f"- {region}")
            else:
                print(f"\nНе знайдено регіонів, де народжуваність у {YEAR_COLUMN} році була вище середньої.")

        else:
            print(f"\nПомилка: Стовпець '{YEAR_COLUMN}' відсутній у таблиці після обробки.")
            print(f"Наявні стовпці: {target_df.columns.tolist()}")

    else:
        print(f"Помилка: Не вдалося знайти таблицю з очікуваними стовпцями, включаючи '{YEAR_COLUMN}', на сторінці.")
        print(f"Всього знайдено таблиць: {len(dfs)}")

except Exception as e:
    print(f"Сталася помилка під час читання або обробки сторінки: {e}")
```
