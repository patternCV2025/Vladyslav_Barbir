```python
import pandas as pd

file_path = "2017_jun_final.csv"
column_to_keep = "Язык.программирования" 

try:
    df = pd.read_csv(file_path)

    print(f"Файл '{file_path}' успішно прочитано.")
    print(f"Початкова кількість стовпців: {df.shape[1]}")

    columns_with_nan_before = df.columns[df.isnull().any()].tolist()
    print(f"\nСтовпці з пропусками ДО видалення: {columns_with_nan_before}")

    columns_to_drop = []
    if column_to_keep not in df.columns:
        print(f"\nПопередження: Стовпець '{column_to_keep}', який потрібно зберегти, відсутній.")
        columns_to_drop = columns_with_nan_before
    else:
        for col in columns_with_nan_before:
            if col != column_to_keep:
                columns_to_drop.append(col)

    print(f"\nСтовпці, що будуть видалені: {columns_to_drop}")

    df_cleaned = df.drop(columns=columns_to_drop)

    print("\nСтовпці успішно видалено.")
    print(f"Кінцева кількість стовпців: {df_cleaned.shape[1]}")
    print(f"Кінцеві стовпці: {df_cleaned.columns.tolist()}")

    print("\nЧастка пропусків у залишених стовпцях:")

    if not df_cleaned.empty:
        nan_counts_after = df_cleaned.isnull().sum()
        total_rows_after = len(df_cleaned)
        nan_proportion_after = nan_counts_after / total_rows_after
        print(nan_proportion_after)

        print("\nПеревірка:")
        columns_with_nan_after = df_cleaned.columns[df_cleaned.isnull().any()].tolist()

        if not columns_with_nan_after:
            print("У жодному із залишених стовпців немає пропусків.")
        elif len(columns_with_nan_after) == 1 and columns_with_nan_after[0] == column_to_keep:
            print(f"Пропуски залишились лише у стовпці '{column_to_keep}', як і очікувалося.")
        else:
            print(f"УВАГА: Пропуски залишились і в інших стовпцях, окрім '{column_to_keep}': {columns_with_nan_after}")

    else:
        print("DataFrame порожній після видалення стовпців.")
    


except FileNotFoundError:
    print(f"Помилка: Файл не знайдено за вказаним шляхом: {file_path}")
    print("Будь ласка, перевірте правильність шляху та наявність файлу.")
except pd.errors.EmptyDataError:
    print(f"Помилка: Файл '{file_path}' порожній.")
except pd.errors.ParserError:
     print(f"Помилка: Не вдалося обробити файл '{file_path}'. Можливо, він має неправильний формат CSV.")
except Exception as e:
    print(f"Сталася невідома помилка: {e}")
```

12. Використовуючи метод `groupby`, групуйте дані по стовпцю `"Должность"`.

```python
import pandas as pd

file_path = "2017_jun_final.csv"
language_column = "Язык.программирования" 
target_language = "Python"
position_column = "Должность" 

try:
    df = pd.read_csv(file_path)

    print(f"Файл '{file_path}' успішно прочитано.")

    if language_column in df.columns and position_column in df.columns:
        
        print(f"\nФільтрація даних: вибір рядків, де '{language_column}' == '{target_language}'...")
        python_mask = df[language_column] == target_language
        python_data = df[python_mask].copy() 

        if not python_data.empty:
             print(f"Створено таблицю 'python_data' розміром: {python_data.shape}")

             print(f"\nГрупування даних таблиці 'python_data' за стовпцем '{position_column}'...")

             grouped_by_position = python_data.groupby(position_column)

             position_counts = grouped_by_position.size()

             print(f"\nКількість Python-спеціалістів за посадою:")
             print(position_counts)

             salary_column = 'Зарплата.в.месяц' 
             if salary_column in python_data.columns:
                  python_data[salary_column] = pd.to_numeric(python_data[salary_column], errors='coerce')
                  avg_salary_by_position = grouped_by_position[salary_column].mean()
                  print(f"\nСередня зарплата Python-спеціалістів за посадою (USD/місяць):")
                  print(avg_salary_by_position.round(2)) 
            
        else:
             print(f"\nТаблиця 'python_data' порожня (не знайдено спеціалістів з {target_language}). Групування неможливе.")

    elif language_column not in df.columns:
         print(f"\nПомилка: Стовпець '{language_column}' не знайдено у файлі.")
    else: 
         print(f"\nПомилка: Стовпець '{position_column}' для групування не знайдено у файлі.")


except FileNotFoundError:
    print(f"Помилка: Файл не знайдено за вказаним шляхом: {file_path}")
    print("Будь ласка, перевірте правильність шляху та наявність файлу.")
except pd.errors.EmptyDataError:
    print(f"Помилка: Файл '{file_path}' порожній.")
except pd.errors.ParserError:
     print(f"Помилка: Не вдалося обробити файл '{file_path}'. Можливо, він має неправильний формат CSV.")
except Exception as e:
    print(f"Сталася невідома помилка: {e}")
```
